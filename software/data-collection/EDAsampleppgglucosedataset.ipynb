```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv("data/raw/Sample_PPG___Glucose_Data.csv")
```

```python
display(df.head())

print("\n\nDataset Info:")
df.info()

print("\nSummary Statistics:")
display(df.describe())


print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Set up the figure
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# Histograms for numerical features
sns.histplot(df["PPG_value"], bins=20, kde=True, ax=axes[0, 0])
axes[0, 0].set_title("PPG Value Distribution")

sns.histplot(df["Blood_Glucose_Level"], bins=20, kde=True, ax=axes[0, 1])
axes[0, 1].set_title("Blood Glucose Level Distribution")

sns.histplot(df["Age"], bins=20, kde=True, ax=axes[0, 2])
axes[0, 2].set_title("Age Distribution")

sns.histplot(df["Weight_lbs"], bins=20, kde=True, ax=axes[1, 0])
axes[1, 0].set_title("Weight Distribution")

sns.histplot(df["Height_cm"], bins=20, kde=True, ax=axes[1, 1])
axes[1, 1].set_title("Height Distribution")

sns.countplot(x=df["Gender"], ax=axes[1, 2])
axes[1, 2].set_title("Gender Distribution")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["PPG_value"], y=df["Blood_Glucose_Level"], hue=df["Gender"], alpha=0.7)
plt.title("PPG Value vs Blood Glucose Level")
plt.xlabel("PPG Value")
plt.ylabel("Blood Glucose Level (mg/dL)")
plt.show()
```
